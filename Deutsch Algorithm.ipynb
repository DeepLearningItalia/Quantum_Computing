```python
from qiskit import *
from qiskit.visualization import plot_histogram
```

```python
# Define the quantum circuit
cbit = ClassicalRegister(1)
qreg = QuantumRegister(2)

# Create the quantum circuit
qc = QuantumCircuit(qreg, cbit)
```

```python
# State Preparation
qc.id(0)
qc.x(1)
qc.barrier()
```

```python
# Step 1
qc.h(0)
qc.h(1)
qc.barrier()
```

```python
# Step 2
qc.cx(0,1) ## example of U_f gate for f(x) constant
qc.barrier()
```

```python
# Step 3 
qc.h(0)
qc.barrier()
```

```python
# Measurement
qc.measure(qreg[0], cbit)

#Draw the quantum circuit
qc.draw('mpl', scale=2)
```

```python
backend = Aer.get_backend('qasm_simulator') # choose how to simulate the circuit

results = execute(qc,backend, shots=1).result().get_counts()
print(results)
plot_histogram(results)
```
